# Polynomial Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
```

```python
dataset.head()
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
x,y
```

### Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

### Training the Polynomial Regression of degree 2 model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,lin_reg.predict(x),c='b')
plt.title("truth or bluff (Linear Regression)")
plt.xlabel("position lavel")
plt.ylabel("Salery")
plt.show()
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,lin_reg_2.predict(x_poly),c='b')
plt.title("truth or bluff (polynomial regression)")
plt.xlabel("position lavel")
plt.ylabel("Salery")
plt.show()
```

### Training the Polynomial Regression of degree 4 model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg_2 = PolynomialFeatures(degree=4)
x_poly_2 = poly_reg_2.fit_transform(x)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(x_poly_2,y)
```

```python
plt.scatter(x,y,c="r")
plt.plot(x,lin_reg_3.predict(x_poly_2))
```

###  Predicting new Result with linear regression

```python
lin_reg.predict([[6.5]])
```

###  Predicting new Result with polynomial regression of degree 2

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

###  Predicting new Result with polynomial regression of degree 4

```python
lin_reg_3.predict(poly_reg_2.fit_transform([[6.5]]))
```
